평균제곱오차(MSE) 오차의 합을 n(독립변수 x 원수의 총 개수)로 나눈 값

```python
import numpy as np

#가상의 기울기 a와 절편 b를 정함
fake_a = 3
fake_b = 76

# X = 공부시간 y = 넘파이 배열
x= np.array([2,4,6,8])
y= np.array([81,93,91,97])

def predict(x):
    return fake_a*x +fake_b

#예측 값이 들어갈 빈 리스트 만듦
predict_result = []

#모든 x 값을 한 번씩 대입해 predict_result 리스트를 완성합니다. 
for i in range(len(x)):
    predict_result.append(predict(x[i]))
    print("공부시간=%.f, 실제점수=%.f, 예측점수=%.f" % (x[i],y[i], predict(x[i])))
    
n= len(x)
def mse(y,y_pred):
    return (1/n)*sum((y-y_pred)**2)
    
print('평균 제곱 오차: ' + str(mse(y, predict_result)))
```

오차를 줄이기 위한 방법
1) 경사 하강법(sgd): 오차이 변화에 따라 이차 함수 그래프를 만들고 학습률(기울기의 부호를 바꾸어 가며 미분 시킴) 설정해 미분 값이 0인 지점을 구함

#오차값을 error로 설정 <br>
error = y-y_pred

```python
#오차 함수를 편미분
a_diff = (2/n)*sum(-x *(error))
b_diff = (2/n)*sum(-(error))
#학습률을 0.03으로 설정한 후 ,학습률을 적용한 a, b값을 다시 산출한다. 
lr = 0.03
a= a-lr*a_diff
b= b-lr*b_diff
```

퍼셉트론: 연결장치로써, 입력 값에 가중치를 조절할 수 있게 만들어 반복학습 실시

퍼셉트론의 한계: XOR문제(둘 중 하나만 1일 때 1 출력됨)

다층 퍼셉트론: 퍼셉트론을 각각 처리하는 은닉층을 만들어 XOR문제를 해결할 수 있음
    입력값 (x1,x2)에 가중치 (w)을 곱하고 바이어스(b)를 더하여 은닉층으로 전송,은닉층의 값을 활성화 함수를 통해 산출해냄

```python
#가중치와 바이어스 선언(행렬 개념)

import numpy as np

w11 = np.array([-2,-2])
w12 = np.array([2,-2])
w2 = np.array([1,1])
b1=3
b2=-1
b3=-1
```

```python
#퍼셉트론 함수 제작
def MLP(x,w,b):
    y= np.sum(w*x)+b
    if y<=0:
        return 0
    else:
        return 1
    
#NAND 게이트(하나라도 0이면 1)
def NAND(x1,x2):
    return MLP(np.array([x1,x2]),w11,b1)

#OR 게이트(두 개 중 하나라도 1이면 1)
def OR(x1,x2):
    return MLP(np.array([x1,x2]),w12,b2)

#AND 게이트(두 개 모두 1이면 1)
def AND(x1,x2):
    return MLP(np.array([x1,x2]),w2,b3)

#XOR 게이트(하나만 1일 떄 1)
def XOR(x1,x2):
    return AND(NAND(x1,x2),OR(x1,x2))
```
